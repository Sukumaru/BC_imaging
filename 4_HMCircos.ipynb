```r
library(NMF)
library(circlize)
library(BBmisc)
```

## Data Loading

```r
nmf_brunet_0_runs30_estim.r = readRDS(file = "./rds/NMFResult.rds")
```

```r
grade_dist = readRDS(file = "./rds/grade_dist.rds")
clin_var_dist = readRDS(file = "./rds/clin_var_dist.rds")
clin_sub_dist = readRDS(file = "./rds/clin_sub_dist.rds")
D_colnorm = readRDS(file = "./rds/D_colnorm.rds")
p_order_numeric_survdec = readRDS(file = "./rds/p_order_numeric_survdec.rds")
feat_dendo = readRDS(file = "./rds/feat_dendo.rds")
feature_annotations = readRDS(file = "./rds/feature_annotations.rds")
load(file = "./rds/clus.Rdata")
```

## Basis Map

```r
## Basis Map
### important
basismap(nmf_brunet_0_runs30_estim.r$fit$`7`,color = "-RdBu", scale="row",Colv=NA,distfun='pearson', hclustfun ="complete")
```

## HeatMaps

**The export dimensions were enlarged to make Figure 4a.**

```r
#### heatmap_grade
#### "#E41A1C","#FF7F00","#999999"
par(mar=c(0,0,0,0))
aheatmap(as.matrix(grade_dist[,-1]), color = "Greys", Colv = NA, Rowv = NA, scale = "c1",labRow=c('Tumor Grade I','Tumor Grade II','Tumor Grade III'))
```

```r
#### heatmap_ERPRHER2
aheatmap(as.matrix(clin_var_dist[,-1]), color = "Greens", Colv = NA, Rowv = NA, scale = "c1",labRow=c('ER positive','ER negative','PR positive','PR negative',"HER2 positive","HER2 negative"))
```

```r
#### heatmap_clinical_subtype
aheatmap(as.matrix(clin_sub_dist[,-1]), color = "-RdBu", Colv = NA, Rowv = NA, scale = "c1",labRow=c('Luminal A','Luminal B','TNBC','HER2-Enriched'))
```

```r
aheatmap(t(normalize(D_colnorm)),color = "-RdBu:50", scale="row",Colv=p_order_numeric_survdec, Rowv = rev(feat_dendo$rowInd), breaks=0, labRow = feature_annotations, legend=FALSE)
```
